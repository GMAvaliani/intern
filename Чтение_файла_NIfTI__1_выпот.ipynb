Аффинная матрица (Affine Matrix) — это математический инструмент, используемый для выполнения аффинных преобразований в пространстве. Аффинные преобразования — это комбинация линейных преобразований (таких как масштабирование, вращение и сдвиг) и переноса. Они обладают свойством сохранения пропорций между точками, но могут изменять углы и длины.

В контексте медицинских изображений (как файлы .nii в формате NIfTI), аффинные матрицы используются, чтобы преобразовать координаты в индексном пространстве изображения (где координаты представляют собой индексы в многомерном массиве данных) в физическое пространство (обычно в миллиметрах или другой единице измерения).

Компоненты аффинной матрицы:
1. Линейные преобразования:
Это верхний левый 3x3 подмассив матрицы. Он отвечает за масштабирование, вращение и искажение.

2. Перенос:
Это верхний правый 3x1 подмассив матрицы. Он отвечает за смещение изображения в пространстве.

3. Перспектива (чаще всего не используется в медицинской визуализации):
Это нижний левый 1x3 подмассив.

4. Домножающий элемент:
Обычно это число 1, расположенное в нижнем правом углу матрицы.

```python
# ПРИМЕР
"""
| a  b  c  tx |
| d  e  f  ty |
| g  h  i  tz |
| 0  0  0  1  |


Здесь:

a, b, c, ... i представляют линейные преобразования.
tx, ty, tz представляют переносы по осям x, y и z соответственно

"""
```

**Применение в медицинской визуализации:**
При работе с медицинскими изображениями, такими как MRI или CT, важно знать не только интенсивность каждого пикселя или вокселя, но и его физическое положение в пространстве, относительно других частей тела или изображения. Именно здесь и используется аффинная матрица: она указывает, как интерпретировать положение каждого вокселя в физическом пространстве.

В общем, аффинная матрица предоставляет мост между "индексным" пространством изображения (где каждый воксель имеет индекс) и "физическим" пространством (где каждый воксель имеет координаты x, y, z в мм или другой единице измерения).

```python
import numpy as np
import matplotlib.pyplot as plt
import nibabel as nib

# Загрузите изображение
img = nib.load('LUNG1-001_effusion_first_reviewer.nii.gz')
img_data = img.get_fdata()

# Функция для преобразования индексов изображения в физические координаты
def index_to_physical(idx, affine):
    idx = np.array(idx + [1])  # Добавляем однородную координату
    phys = np.dot(affine, idx)
    return phys[:-1]  # Убираем однородную координату

# Функция для преобразования физических координат в индексы изображения
def physical_to_index(phys, affine):
    phys = np.array(list(phys) + [1])
    idx = np.linalg.inv(affine).dot(phys)
    return tuple(idx[:-1].astype(int))  # Убираем однородную координату и конвертируем в кортеж

# Ваша аффинная матрица
affine = img.affine

# Визуализируем срез изображения в физическом пространстве
slice_idx = img_data.shape[2] // 2
plt.imshow(img_data[:, :, slice_idx], cmap='gray')
plt.title(f"Slice at z-index: {slice_idx} (physical z: {index_to_physical([0, 0, slice_idx], affine)[2]:.2f})")
plt.colorbar()
plt.show()
```

```python
import numpy as np
import nibabel as nib
from scipy.ndimage import zoom
import matplotlib.pyplot as plt

# Функция ресемплирования
def resample(image, scan, new_spacing=[1, 1, 1]):
    # Определение текущего интервала пространственной сетки
    spacing = np.array(scan.header.get_zooms())

    # Вычисление коэффициентов изменения размера
    resize_factor = spacing / new_spacing

    # Изменение размера изображения
    resampled_image = zoom(image, resize_factor, mode='nearest')

    return resampled_image

# Загрузка файла NIfTI
nii_img = nib.load('LUNG1-001_effusion_first_reviewer.nii.gz')

# Извлечение данных изображения и метаданных
img_data = nii_img.get_fdata()

# Выполнение ресемплирования
resampled_img = resample(img_data, nii_img)

# Коррекция данных после ресемплирования
# 1. Обнуление отрицательных значений
resampled_img[resampled_img < 0] = 0

# 2. Обрезка (Clamping)
resampled_img = np.clip(resampled_img, 0, img_data.max())

# 3. Нормализация
resampled_img = (resampled_img - resampled_img.min()) / (resampled_img.max() - resampled_img.min())
resampled_img = resampled_img * (img_data.max() - img_data.min()) + img_data.min()

# Уменьшение объема данных
subsample_rate = 5  # берем каждый пятый пиксель
reduced_img = resampled_img[::subsample_rate, ::subsample_rate, ::subsample_rate]

# Упрощенная визуализация
# Визуализация центрального среза по третьему измерению (z)
plt.imshow(reduced_img[:, :, reduced_img.shape[2] // 2], cmap='gray')
plt.title('Central Slice of the Resampled Image')
plt.colorbar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import measure

def plot_3d(image, threshold=0.2):
    """Визуализация 3D Nifti изображения с использованием marching cubes."""

    # Используйте marching cubes для извлечения поверхности
    verts, faces, _, _ = measure.marching_cubes(image, threshold)

    fig = plt.figure(figsize=(15, 7))

    # Первый график: поворот по оси Z
    ax1 = fig.add_subplot(121, projection='3d')
    ax1.plot_trisurf(verts[:, 0], verts[:, 1], faces, verts[:, 2],
                     cmap='copper', edgecolor='none')
    ax1.set_xlabel("X-axis")
    ax1.set_ylabel("Y-axis")
    ax1.set_zlabel("Z-axis")
    ax1.set_title(f"3D Visualization with threshold: {threshold}")

    # Второй график: поворот по оси Y
    ax2 = fig.add_subplot(122, projection='3d')
    ax2.plot_trisurf(verts[:, 0], verts[:, 1], faces, verts[:, 2],
                     cmap='copper', edgecolor='none')
    ax2.set_xlabel("X-axis")
    ax2.set_ylabel("Y-axis")
    ax2.set_zlabel("Z-axis")
    ax2.view_init(elev=0, azim=90)  # Поворот по оси Y
    ax2.set_title(f"3D Visualization with threshold: {threshold}\nRotation: Y-axis")

    plt.tight_layout()  # Улучшает расположение графиков на изображении
    plt.show()

# Ваши данные
img_data = img.get_fdata()

# Визуализация
plot_3d(img_data, threshold=0)
```
